```python
new_df = pd.read_csv('../Data/heart-disease-cleaned.csv')
```

### Looking at the correlation matrix, we can see that the highest positive correlations with the target are Chest Pain (cp), Exercise Induced Angina (exang), ST depression induced by exercise relative to rest (oldpeak),  number of major vessels (0-3) colored by flouroscopy (ca), and Thalium heart scan (thal). Maximum heart rate achieved (thalach) has a high negative correlation

```python
new_df.corr()['diagnosis']
```

## This is easier to see here

```python
new_df.corr()[(new_df.corr() > 0.4) | (new_df.corr() < -0.4)].fillna('')['diagnosis']
```

## Digging deeper with scaled and dummied data, we can more accurately point out correlations of specific dummy variables.

```python
from eda import get_scaled_dummied_corr
scaled_corr_df = get_scaled_dummied_corr(df)
```

```python
scaled_corr_df['diagnosis']
```

### Here we can see the inidividual values of the categorical predictors that have the highest correlation with the target variable

#### In Chest Pain, the main predictor is Asymptomatic chest pain, with the thalium heart scan, it shows that the Reversable Defect has a high positive correlation, while a Normal Thalium heart scan has a negative correlation

```python

scaled_corr_df[(scaled_corr_df > 0.4) | (scaled_corr_df < -0.4)].fillna('')['diagnosis'].drop('diagnosis', axis=0)
```
